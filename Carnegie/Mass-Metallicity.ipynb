```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
observations=h5py.File('massMetallicityRelationBlanc2019.hdf5','r')
```

```python
#Galacticus file
file=h5py.File('emissionLines_1000_mass3e13_SFR_MresScaled2-5e9_bpt_z0_Gunawardhana2013_vdisSph_mY0-04_R0-6.hdf5','r')
# emissionLines_1000_mass3e13_SFR_MresScaled2-5e9_bpt_z0_Gunawardhana2013_vdisSph_mY0-04_R0-5.hdf5
#merger tree weight
w=file['Outputs']['Output1']['nodeData']['mergerTreeWeight'][:]
```

```python
observations.keys()
```

```python
list(file['Outputs']['Output1']['nodeData'].keys())
```

```python
Metals_gas=file['Outputs']['Output1']['nodeData']['diskAbundancesGasMetals'][:]+file['Outputs']['Output1']['nodeData']['spheroidAbundancesGasMetals'][:]
Zpsun=0.0198

H_gas=file['Outputs']['Output1']['nodeData']['diskMassGas'][:]+file['Outputs']['Output1']['nodeData']['spheroidMassGas'][:]
mass_stellar=file['Outputs']['Output1']['nodeData']['diskMassStellar'][:]+file['Outputs']['Output1']['nodeData']['spheroidMassStellar'][:]
```

```python
Zp=(Metals_gas/H_gas)/Zpsun

Z=np.log10(Zp)+8.68
```

```python
SFR=file['Outputs']['Output1']['nodeData']['spheroidStarFormationRate'][:]+file['Outputs']['Output1']['nodeData']['diskStarFormationRate'][:]
```

```python
print(SFR[SFR>2E8])
```

```python
O_mean=observations['abundanceOxygenMean'][:]
O16=observations['abundanceOxygen16PercentCI'][:]
O84=observations['abundanceOxygen84PercentCI'][:]
massStellar=observations['massStellar'][:]
```

```python
print(O84,O_mean)
```

```python
print(observations['abundanceOxygen16PercentCI'][:])
```

```python
H_gasd=file['Outputs']['Output1']['nodeData']['diskMassGas'][:]
H_gass=file['Outputs']['Output1']['nodeData']['spheroidMassGas'][:]

mass=file['Outputs']['Output1']['nodeData']['basicMass'][:]

mask=np.isnan(Z) 
mask1=np.isinf(Z[~mask])
masksfr=SFR[~mask][~mask1]>1e8
satellite=file['Outputs']['Output1']['nodeData']['nodeIsIsolated'][:][~mask][~mask1][masksfr]==0
```

```python
plt.scatter(H_gas[~mask][~mask1][masksfr][satellite],Z[~mask][~mask1][masksfr][satellite],c='b',s=10,alpha=.5,label='satellites')
# plt.scatter(H_gasd[~mask][~mask1][masksfr][~satellite],Z[~mask][~mask1][masksfr][~satellite],c='g',s=10,alpha=.5,label='disk')
plt.scatter(H_gas[~mask][~mask1][masksfr][~satellite],Z[~mask][~mask1][masksfr][~satellite],c='gold',s=10,alpha=.5,label='Centrals')


plt.ylim(8,9.5)
plt.xlim(1e6,1e11)
plt.legend()
plt.xlabel('Gas mass (disk+sph)')
plt.ylabel(r'$12\,+\,\mathrm{log}(\mathrm{O/H})\,(|Z|)$')
plt.xscale('log')
```

```python
plt.scatter(mass[~mask][~mask1][masksfr],Z[~mask][~mask1][masksfr],c='k',s=10,alpha=.5)
# plt.scatter(H_gasd[~mask][~mask1][masksfr][~satellite],Z[~mask][~mask1][masksfr][~satellite],c='g',s=10,alpha=.5,label='disk')
# plt.scatter(H_gas[~mask][~mask1][masksfr][~satellite],Z[~mask][~mask1][masksfr][~satellite],c='gold',s=10,alpha=.5,label='Centrals')


plt.ylim(8,9.5)
plt.xlim(1e9,1e13)
plt.legend()
plt.xlabel('Halo mass')
plt.ylabel(r'$12\,+\,\mathrm{log}(\mathrm{O/H})\,(|Z|)$')
plt.xscale('log')
```
